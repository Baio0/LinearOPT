```python
def row_to_numpy_2d_P(row, cat):
    data={}
    row2=row.copy()
    label =row2['label']
    del row2['label']
    if cat != label : 
        return False, None,None,label

    N=len(row2)
    np_array=np.array(list(row2.values()),dtype=np.float64)
    np_array=np_array.reshape((int(np_array.shape[0]/3),3))
    X=np_array[:,[0,1]]
    p=np_array[:,2]
    D=p>0

    X_take=X[D]
    p_take=p[D]
    p_take=p_take/np.sum(p_take)
    data['X']=X_take
    data['p']=p_take
    return True, X_take,p_take, label


# def row_to_numpy_2d_P(row, cat):
#     data={}
#     row2=row.copy()
#     data['label'] =row2['label']
#     del row2['label']
#     N=len(row2)
#     np_array=np.array(list(row2.values()),dtype=np.float64)
#     np_array=np_array.reshape((int(np_array.shape[0]/3),3))
#     X=np_array[:,[0,1]]
#     p=np_array[:,2]
#     D=p>0
#     if cat != data['label'] : 
#         return False, data
#     X_take=X[D]
#     p_take=p[D]
#     data['X']=X_take
#     data['p']=p_take
#     return True, data
# data=row_to_numpy(row)
# print(data['X'][0])
```

```python
#load and convert data 
import numpy as np
import csv

X_list = []
p_list = []
label_list=[]
cat_list = [str(i) for i in range(10)]
for cat in cat_list:
    label_counts=0
    with open('experiment/data/MNIST/train.csv', newline='') as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            cond, X, p, label = row_to_numpy_2d_P(row, cat)
            if cond:
                X_list.append(X)
                p_list.append(p)
                label_list.append(label)
                label_counts+=1
            #print(len(X_list))
            if label_counts==300:
                break


# X_list = np.array(X_list)
# p_list = np.array(p_list)


# X_list = np.array(X_list)
# p_list = np.array(p_list)
# label_list = np.array(label_list)



        
file = (X_list,p_list, label_list)
torch.save(file,'experiment/data/MNIST/pca.pt')
        

#             break;
```
